```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re

%matplotlib inline
```

```python
filepath = './contagens/itapoa-01.txt'
    
df = pd.read_csv(filepath, names=['Distance'])
```

```python
# Cleaning and transforming

# Changing idle readings to 0
df[df['Distance'] > 8189] = 0
df.describe()
```

```python
plt.rcParams['figure.figsize'] = (18,9)

# Plot entire dataset
plt.plot(df['Distance'])
plt.title('Todas as leituras')
plt.xlabel('Tempo (ms)')
plt.ylabel('Distncia')
plt.grid(True)
```

```python
# Set window for pontual analysis
win_size = 100
win_start = 500
win_end = win_start + win_size

# Ploting window entries
wdf = df.iloc[win_start:win_end].copy()
plt.plot(wdf['Distance']);
plt.title('Leituras de %d a %d' % (win_start, win_end))
plt.xlabel('Tempo (ms)')
plt.ylabel('Distncia')
plt.grid(True)
```
